- 다음 요구조건에 따라 시각화를 진행하시오.
    - Genotype의 분포를 히스토그램과 파이 차트로 나타내고, 이를 한 그래프에 포함시키시오. (subplot 이용)
    - 모든 그래프에 적절한 그래프 이름과 축 이름을 추가하시오.
    - 모든 그래프를 회색조로 그리시오.

```python
# 파일 다운로드
file_id = '11sBXR59oIszhfPwHxaFEz4mFyFjt_zp1'
url = f'https://drive.google.com/uc?id={file_id}'

output = 'genome.csv'
gdown.download(url, output, quiet=False)

genome = pd.read_csv(output)
```

```python
# 히스토그램
plt.subplot(1,2,1)
plt.hist(genome['genotype'], color='grey', edgecolor='black')
plt.title('Number of Genomes per Genotype')
plt.xlabel('genotype')
plt.ylabel('number')

# 파이차트
## percentage 계산
gt_list = ['AA','GG','AG','CC','CT','GT','TT','AC','AT','CG']
percentage = []
idx = 0

for i in range(10):
    percentage.append(((genome['genotype'].value_counts()[gt_list[idx]])
                  /(len(genome['genotype'])))*100)
    idx = idx+1
    
##그래프
plt.subplot(1,2,2)
plt.pie(percentage, labels = gt_list, autopct='%.1f%%', startangle=120, counterclock=True, 
        colors=['#222222','#333333','#444444','#555555','#666666','#777777','#888888','#999999','#aaaaaa','#bbbbbb'],)
plt.title('Percentage of Genomes per Genotype')
```
