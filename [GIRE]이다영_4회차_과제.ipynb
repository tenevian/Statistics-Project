```python
# 파일 다운로드
file_id = '1SkCbZdnzng1a43m_vRRrVbmeFOPcyIU0'
url = f'https://drive.google.com/uc?id={file_id}'

output = 'cancer.csv'
gdown.download(url, output, quiet=False)

# cancer 변수에 저장
cancer = pd.read_csv(output)
```

```python
cancer.head()
```

```python
# 주 이름을 인덱스로 사용
cancer_idx = cancer.set_index('State')
cancer_idx.head()
```

```python
# 결측치 확인
cancer_idx.isnull().any()
```

```python
# 결측치를 평균값으로 채우기
cancer_idx.fillna(np.mean(cancer['Rate']))
```

```python
# 이상치를 IQR 방식으로 처리

#아래 코드는 AI 도움 받아서 작성된 것인데 원본코드처럼 에러가 뜨진 않지만 목표한 결과가 나오지도 않는 것 같아서 코드를 어떻게 고쳐야 할지 잘 모르겠습니다

def iqr_outlier(data):
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            q1, q3 = np.percentile(data[column], [25, 75])
            iqr = q3 - q1
            lower = q1 - (iqr * 1.5)
            upper = q3 + (iqr * 1.5)

            print(f"Column: {column}")
            print(f"q1: {q1}")
            print(f"q3: {q3}")
            print(f"IQR: {iqr}")
            print(f"lower: {lower}")
            print(f"upper: {upper}")

            data = data[(data[column] >= lower)]
            data = data[(data[column] <= upper)]

    data = data.reset_index(drop=True)
    return data

iqr_outlier(cancer_idx)
```

```python
# IQR 처리 수업자료 원본 함수

def iqr_outlier(data):
    for i in range(len(data.iloc[3])):
        q1, q3 = np.percentile(data.iloc[:, i], [25, 75])
        # percentile 함수
        print(f"q1: {q1}") #percentile 값 출력
        print(f"q3: {q3}")

        iqr = q3 - q1 #사이 간격
        lower = q1 - (iqr * 1.5)
        upper = q3 + (iqr * 1.5)
        print(f"lower: {lower}") #bound 표시
        print(f"upper: {upper}")

        data = data[data.iloc[:, i] < upper]
        data = data[data.iloc[:, i] > lower]
        
    data = data.reset_index(drop=True)
    return data
```
