```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

# 2회차 과제
목표: 빅데이터 기본 처리

다음은 분꽃에 관한 여러 데이터이다. Python을 이용해 적절히 분석하시오.

```python
iris_raw = load_iris()
iris = pd.DataFrame(data=iris_raw.data, columns=iris_raw.feature_names)
display(iris)
```

```python
# 꽃받침(sepal)의 평균 길이를 구하시오.
print(np.mean(iris['sepal length (cm)']))
```

```python
# 꽃잎(petal)의 최대 길이와 최소 길이를 구하시오.
print(np.min(iris['petal length (cm)']))
print(np.max(iris['petal length (cm)']))
```

```python
# 꽃받침의 길이와 너비가 각각 5cm, 3cm 이상인 데이터만을 선별하시오.
iris[(iris['sepal length (cm)'] >= 5) & (iris['sepal width (cm)'] >= 3)]
```

```python
# 꽃잎의 너비가 0.2 이하인 데이터는 총 몇 개인가?


pw = iris[iris['petal width (cm)'] <= 0.2]
print(len(pw)+1)
```

```python
# 꽃잎의 너비, 꽃받침의 너비, 꽃잎의 길이, 꽃받침의 길이 순서로 내림차순 정렬하시오.
iris.sort_values(by=['petal width (cm)','sepal width (cm)','petal length (cm)','sepal length (cm)'], 
                 ascending = [False, False, False, False])
```
