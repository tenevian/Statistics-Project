### 정규 분포의 활용
- 최근 5년간 통계학 시험 결과의 평균은 70점이고, 표준 편차는 10점인 것으로 나타났다.
- 이를 그래프로 나타내시오.

```python
from scipy import stats
rv = stats.norm(70,10) #표준편차 만들기
x = np.linspace(0, 100, 1000) #데이터 만들기: x값 설정
y1 = rv.pdf(x) #y값 설정
plt.plot(x,y1) #plot points
plt.show() #display graph
```

- 올해 시험의 난이도가 상승해 평균은 10점 낮아지고 표준편차는 5점 증가했다.
- 이를 그래프로 나타내시오.

```python
from scipy import stats
rv = stats.norm(60,15)
x = np.linspace(0, 100, 1000)
y2 = rv.pdf(x)
plt.plot(x,y2)
plt.show()
```

- 그래프의 형태에 어떤 차이가 있는지 간략하게 기술하시오.

답: 
- 두 번째 그래프에서 최댓값이 낮아지고 그래프가 가로로 넓어진 모양이 되었다. 
- 그래프의 꼭짓점에 해당하는 x 값(평균값)이 70에서 60으로 낮아졌다.
- 표준 편차가 증가하며 데이터 값이 더 고르게 퍼졌다. 

### 표본 분포의 활용
- 경기도 지역 낙농 목장의 젖소가 만드는 평균 우유의 양을 알아보기 위해 목장 10개소에 대한 평균 우유의 양을 조사하였다.
- 그 결과, 평균이 35 kg이고 표준편차가 5 kg으로 나타났다.
- 이때, 적절한 **표준** T 분포를 그리시오.

```python
from scipy import stats

rv = stats.t(9)

x = np.linspace(-5, 5, 100)
y = rv.pdf(x)

plt.plot(x,y)
plt.show()
```

- 목장 100개소를 조사했을 때 적절한 T 분포를 그리시오.

```python
from scipy import stats

rv = stats.t(99)

x = np.linspace(-5, 5, 100)
y = rv.pdf(x)

plt.plot(x,y)
plt.show()
```

- 두 그래프와 표준정규분포 그래프를 동시에 나타낸 후 비교하시오.

```python
from scipy import stats

x = np.linspace(-5, 5, 100)

for i in [9,99]:
    rv = stats.t(i)
    y = rv.pdf(x)
    plt.plot(x,y)
    
rv = stats.norm(0, 1)
y = rv.pdf(x)
plt.plot(x, y)


plt.legend(["First T-dist.", "Second T-dist.", "Std. Normal Dist."])
plt.show()
```

### 비율 분포
- F 분포에 대하여, 자유도가 (3, 5) 일 때 그래프를 그리시오.

```python
rv = stats.f(3,5)
x = np.linspace(0, 5, 1000)
y = rv.pdf(x)
plt.plot(x,y)
plt.xlim(left = 0)
plt.ylim(bottom = 0)
plt.show()
```
